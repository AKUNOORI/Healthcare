```python
from pyspark.sql.functions import col

df = spark.table("healthcareNew.bronze.ChatHistory")\
    .select ("ChatID", "ClassID","ChatMessage", "FlaggedWords", "IsFlagged", "LanguageFlag", "SendToProfileID", "AdminUserID", "StudentProfileID", "CreatedTimestamp")\
        .filter(col("ClassID").isNotNull() & col("ChatID").isNotNull())
df.write.format("delta").mode("overwrite").saveAsTable("healthcareNew.silver.ChatHistory")
print("silver.ChatHistory written")
```

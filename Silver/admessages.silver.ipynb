```python
from pyspark.sql.functions import col

df = spark.table("healthcareNew.bronze.admessages")\
    .select("ADMessageID","AdCaptureinformationLine1","AdCaptureinformationLine2",\
        "AdCaptureinformationLine3","AdCaptureinformationLine4","AdResponseMessage",\
            "CreatedBy","EnabledFlag","EndDateActive","MessageText","ModifiedBy","ModifiedDate")\
                .filter(col("EnabledFlag") == "Y")
df.write.format("delta").mode("overwrite").saveAsTable("healthcareNew.silver.admessages")
```

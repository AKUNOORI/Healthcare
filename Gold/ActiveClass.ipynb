```python
%sql
CREATE OR REPLACE TABLE healthcarenew.gold.ActiveClassByStudents
USING DELTA
AS
SELECT 
    sp.StudentProfileID,
    cs.ClassID,
    sp.UserName,
    cs.FirstName,
    cs.LastName,
    cs.iPadNumber,
    CASE 
        WHEN cs.BlockChat = True THEN 'Yes' 
        ELSE 'No' 
    END AS BlockChatStatus,
    CASE 
        WHEN cs.BlockGame = True THEN 'Yes' 
        ELSE 'No' 
    END AS BlockGameStatus
FROM healthcarenew.silver.classstudent cs
JOIN healthcarenew.silver.studentprofile sp
  ON cs.StudentProfileID = sp.StudentProfileID
WHERE cs.ClassID = TRY_CAST(:classid AS BIGINT)
   OR :classid IS NULL
   OR :classid = ''
```
